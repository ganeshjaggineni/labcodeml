```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=np.array([23,42,12,10,15,14,9])
print(data.shape)
print(data)
```

```python
feature_names=['x']
df=pd.DataFrame(data,columns=feature_names)
print(df)
```

```python
max_val = df['x'].max()
min_val = df['x'].min()
median_val= df['x'].median()
q1 = df['x'].quantile(0.25)
q2= df['x'].quantile(0.50)
q3 = df['x'].quantile(0.75)
print(max_val)
print(min_val)
print(median_val)
print(q1)
print(q2)
print(q3)
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(df['x'],orient='V')
```

```python
df.hist(bins=12)
plt.suptitle("histogram")
```

```python
sns.histplot(df['x'])
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(df['x'],orient='v')
plt.text(0.85,q1,f'q1={q1:.2f}',ha='left',va='center',fontweight='bold')
plt.text(0.85,q2,f'q2={q2:.2f}',ha='left',va='center',fontweight='100')
```
