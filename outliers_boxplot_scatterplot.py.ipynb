```python
import pandas as pd
import numpy as np

arr = np.array([[3, 0, 12, 0, 2, 0, 26, 0, 7, 5, 5, 2, 1, 1, 2],
                [12, 0, 9, 4, 15, 2, 6, 10, 27, 15, 5, 9, 1, 14, 2]])

df = pd.DataFrame(arr.T, columns=['x', 'y'])

# Transpose the array (arr.T) to match the columns correctly

print(df)
```

```python
def find_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    print(Q1)
    print(Q3)
    print(IQR)
    print(lower_bound)
    print(upper_bound)
    outliers = []
    for value in column:
        if value > upper_bound or value < lower_bound:
            outliers.append(value)
    return outliers
    
outliers_x=find_outliers(df['x'])
print("\n")
outliers_y=find_outliers(df['y'])
```

# DRAWING OUTLIERS USING SCATTERPLOT
plt.figure(figsize(10,8))

```python
plt.figure(figsize=(10,8))
plt.scatter(df['x'],range(len(df['x'])),label="datapoints",color="blue")
plt.scatter(outliers_x,range(len(outliers_x)),label="outliers",color="red")
plt.legend()
```
