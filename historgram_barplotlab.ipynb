# IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# DATASET

```python
bulb_life=["300-400","400-500","500-600","600-700","700-800","800-900","900-1000"]
no_of_lamps=[14,56,60,86,74,62,48]
data = pd.DataFrame({'Bulb Life': bulb_life, 'Number of Lamps': no_of_lamps})
print(data)
```

```python
plt.figure(figsize=(10,8))
plt.bar(bulb_life,no_of_lamps,color="blue",edgecolor="black")
plt.xlabel("bulb_life")
plt.ylabel("no_of_lamps")
plt.title('Bulb Life Intervals vs. Number of Lamps')
plt.xticks(rotation=45)# Rotate x-axis labels for better readability
plt.show()
```
